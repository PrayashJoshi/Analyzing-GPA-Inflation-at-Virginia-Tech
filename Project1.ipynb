<h2>Project introduction:</h2>


<h3>Intro Description:</h3>


VT has a large dataset of course grade distributions dating back to 2012, we plan to visualize the data and draw conclusions/correlations about courses that could be relevant for evaluation by students and/or administrators.

<h3>Potential research questions you plan to address:</h3>




* Look at trends over the past 10 years for grades given a course. 
* Analyze and extract the professors with the highest percent of “A”, professors who have the largest difference in percent of “A” compared to the next highest and various other statistics that students would be interested in when deciding to take a course. 
* Determine which courses are distributed the closest to a standard or ideal grade distribution.
